```python
import tensorflow as tf
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
 
tf.enable_eager_execution()
 
 
image_path = '/Users/nikhil/Downloads/attachments 2/image/20588216_image.png'
 
image_string=tf.read_file(image_path)
 
image=tf.image.decode_jpeg(image_string,channels=3)
# image=tf.image.convert_image_dtype(image,dtype=tf.float32)
 
 
def show_image(original_image,augmented_image,title):
  fig=plt.figure()
  fig.suptitle(title)
  
  original_plt=fig.add_subplot(1,2,1)
  
  original_plt.set_title('original image')
  original_plt.imshow(original_image)
 
  augmented_plt=fig.add_subplot(1,2,2) 
  augmented_plt.set_title('augmented image')
  augmented_plt.imshow(augmented_image)
  plt.show(block=True)
```

## Brightness

```python
delta_1 = tf.image.adjust_brightness(image,0.01)
show_image(image,delta_1,'Brightness-delta 0.01')
```

```python
delta_2 = tf.image.adjust_brightness(image,0.1)
show_image(image,delta_2,'Brightness-delta 0.1')
```

```python
delta_3 = tf.image.adjust_brightness(image,0.5)
show_image(image,delta_3,'Brightness delta 0.5')
```

## Contrast

```python
contrast_1 = tf.image.adjust_contrast(image,contrast_factor=0.01)
show_image(image,contrast_1,'Brightness contrast 0.01')
```

```python
contrast_2 = tf.image.adjust_contrast(image,contrast_factor=0.1)
show_image(image,contrast_2,'Brightness contrast 0.1')
```

```python
contrast_3 = tf.image.adjust_contrast(image,contrast_factor=0.5)
show_image(image,contrast_3,'Brightness contrast 0.5')
```

```python
contrast_4 = tf.image.adjust_contrast(image,contrast_factor=1)
show_image(image,contrast_4,'Brightness contrast 1')
```
