```python
from PIL import Image
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

```python
#new = Image.open('test.png')
img = cv.imread('test.png',0)
```

```python
plt.imshow(img,cmap='gray')
```

```python
erosion = cv.erode(img,kernel,iterations = 1)
```

```python
fig, ax = plt.subplots(1,2)
ax[0].set_title("original")
ax[1].set_title("erosion")
ax[0].imshow(img,cmap='gray');
ax[1].imshow(erosion,cmap='gray');
```

```python
kernel = np.ones((5,5),np.uint8)
```

```python
tophat = cv.morphologyEx(img, cv.MORPH_TOPHAT, kernel)
```

```python
fig, ax = plt.subplots(1,2)
ax[0].set_title("original")
ax[1].set_title("tophat")
ax[0].imshow(img,cmap='gray');
ax[1].imshow(tophat,cmap='gray');
```

```python
dilation = cv.dilate(img,kernel,iterations = 1)

fig, ax = plt.subplots(1,2)
ax[0].set_title("original")
ax[1].set_title("dilation")
ax[0].imshow(img,cmap='gray');
ax[1].imshow(dilation,cmap='gray');
```

```python
opening = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)
fig, ax = plt.subplots(1,2)
ax[0].set_title("original")
ax[1].set_title("opening")
ax[0].imshow(img,cmap='gray');
ax[1].imshow(opening,cmap='gray');
```

```python
closing = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)
fig, ax = plt.subplots(1,2)
ax[0].set_title("original")
ax[1].set_title("closing")
ax[0].imshow(img,cmap='gray');
ax[1].imshow(closing,cmap='gray');
```

```python
gradient = cv.morphologyEx(img, cv.MORPH_GRADIENT, kernel)
fig, ax = plt.subplots(1,2)
ax[0].set_title("original")
ax[1].set_title("gradient")
ax[0].imshow(img,cmap='gray');
ax[1].imshow(gradient,cmap='gray');
```

```python
blackhat = cv.morphologyEx(img, cv.MORPH_BLACKHAT, kernel)
fig, ax = plt.subplots(1,2)
ax[0].set_title("original")
ax[1].set_title("blackhat")
ax[0].imshow(img,cmap='gray');
ax[1].imshow(blackhat,cmap='gray');
```
